## 💗 Chapter 2 — The Heart Equation

The parametric heart curve is defined as:

$$x(t) = 16 \sin^3(t)$$
$$y(t) = 13\cos(t) - 5\cos(2t) - 2\cos(3t) - \cos(4t)$$

> *Even mathematics has known about love for centuries.* 🌸


```python
t = np.linspace(0, 2 * np.pi, 1000)

# Classic parametric heart
x = 16 * np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)

fig, ax = plt.subplots(figsize=(7, 6.5))
fig.patch.set_facecolor('#fff0f5')
ax.set_facecolor('#fff8fb')

# Gradient-fill the heart
from matplotlib.path import Path
from matplotlib.patches import PathPatch
verts = list(zip(x, y))
codes = [Path.MOVETO] + [Path.LINETO]*(len(verts)-2) + [Path.CLOSEPOLY]
path  = Path(verts, codes)
patch = PathPatch(path, facecolor='#f4a8c3', edgecolor='#d63384', linewidth=2, alpha=0.85)
ax.add_patch(patch)

ax.plot(x, y, color='#d63384', linewidth=2)

# Names inside
ax.text(0, 2,  f'{NAME_1}', ha='center', va='center',
        fontsize=15, color='#fff', fontfamily='DejaVu Serif',
        fontstyle='italic', fontweight='bold')
ax.text(0, -2, f'& {NAME_2}', ha='center', va='center',
        fontsize=15, color='#fff', fontfamily='DejaVu Serif',
        fontstyle='italic', fontweight='bold')
ax.text(0, -6.5, f'💗  {SPECIAL_DATE}  💗', ha='center', va='center',
        fontsize=11, color='#d63384', fontfamily='DejaVu Serif', fontstyle='italic')

ax.set_xlim(-18, 18)
ax.set_ylim(-14, 14)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title('The Heart Curve  —  $x = 16\\sin^3(t),\\; y = 13\\cos t - 5\\cos 2t - 2\\cos 3t - \\cos 4t$',
             fontsize=10, color='#7a4060', pad=12)
plt.tight_layout()
plt.show()
```

## 🌀 Chapter 3 — The Golden Spiral of Love

The **Archimedean spiral** $r = a \cdot \theta$ keeps growing outward, just like love.

The **Golden Ratio** $\phi = \frac{1+\sqrt{5}}{2} \approx 1.618$ appears in nature's most beautiful patterns — and in us. 🌿


```python
phi = (1 + np.sqrt(5)) / 2   # Golden Ratio ≈ 1.618
print(f'Golden Ratio φ = {phi:.6f}')
print(f'φ² = {phi**2:.6f}  (= φ + 1 — it grows into itself)')
print()

theta = np.linspace(0, 6 * np.pi, 1500)

# Two spirals — one for each of us — growing together
r1 = phi * theta           # Nachiket's spiral
r2 = phi * (theta + np.pi) # Kalyani's spiral (offset by π — we complement each other)

fig, ax = plt.subplots(figsize=(7, 7), subplot_kw={'projection': 'polar'})
fig.patch.set_facecolor('#fff0f5')
ax.set_facecolor('#fff8fb')

# Colour maps — pink spectrum
colors1 = plt.cm.RdPu(np.linspace(0.3, 0.95, len(theta)))
colors2 = plt.cm.PuRd(np.linspace(0.3, 0.95, len(theta)))

for i in range(len(theta)-1):
    ax.plot(theta[i:i+2], r1[i:i+2], color=colors1[i], linewidth=1.4, alpha=0.85)
    ax.plot(theta[i:i+2], r2[i:i+2], color=colors2[i], linewidth=1.4, alpha=0.7)

ax.set_yticklabels([])
ax.set_xticklabels([])
ax.grid(color='#f4a8c3', linestyle='--', linewidth=0.4, alpha=0.5)
ax.spines['polar'].set_color('#f4a8c3')

# Legend
p1 = mpatches.Patch(color='#d63384', label=f'{NAME_1} — r = φθ')
p2 = mpatches.Patch(color='#e8729a', label=f'{NAME_2} — r = φ(θ+π)')
ax.legend(handles=[p1, p2], loc='lower right', fontsize=9,
          framealpha=0.7, facecolor='#fff0f5', edgecolor='#f4a8c3')

ax.set_title(f'The Golden Spiral of {NAME_1} & {NAME_2}\n'
             r'$r = \phi\theta\;$ where $\phi = \frac{1+\sqrt{5}}{2}$',
             fontsize=11, color='#7a4060', pad=20)
plt.tight_layout()
plt.show()
print(f'\n💗 Two spirals, offset by π — different, yet perfectly paired.')
```

## 🎨 Chapter 4 — Lissajous Love Curves

**Lissajous figures** are parametric curves:
$$x = A\sin(at + \delta), \quad y = B\sin(bt)$$

> *When two frequencies find harmony, they draw the most beautiful shapes.*
> *Just like us.* 🌸


```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
fig.patch.set_facecolor('#fff0f5')
fig.suptitle(f'Lissajous Love Curves  —  {NAME_1} & {NAME_2}',
             fontsize=14, color='#d63384', fontfamily='DejaVu Serif',
             fontstyle='italic', y=1.01)

configs = [
    (1, 1, np.pi/2, '#e8729a', 'a=1, b=1  — Circle of Trust'),
    (1, 2, np.pi/4, '#d63384', 'a=1, b=2  — Figure Eight'),
    (3, 2, np.pi/4, '#f4a8c3', 'a=3, b=2  — Butterfly'),
    (3, 4, np.pi/6, '#c9506e', 'a=3, b=4  — Crown'),
    (5, 4, np.pi/4, '#e8729a', 'a=5, b=4  — Star'),
    (5, 6, np.pi/3, '#d63384', 'a=5, b=6  — Weave'),
]

t = np.linspace(0, 2*np.pi, 3000)
for ax, (a, b, delta, color, title) in zip(axes.flat, configs):
    ax.set_facecolor('#fff8fb')
    x = np.sin(a * t + delta)
    y = np.sin(b * t)
    # colour-gradient line
    cmap = LinearSegmentedColormap.from_list('love', ['#fce4ec', color])
    points = np.array([x, y]).T.reshape(-1, 1, 2)
    segs = np.concatenate([points[:-1], points[1:]], axis=1)
    from matplotlib.collections import LineCollection
    lc = LineCollection(segs, cmap=cmap, linewidth=1.4, alpha=0.9)
    lc.set_array(np.linspace(0, 1, len(segs)))
    ax.add_collection(lc)
    ax.set_xlim(-1.2, 1.2); ax.set_ylim(-1.2, 1.2)
    ax.set_aspect('equal'); ax.axis('off')
    ax.set_title(title, fontsize=8.5, color='#7a4060')
    # heart in corner
    ax.text(0.92, 0.92, '💗', transform=ax.transAxes, fontsize=9, ha='right', va='top')

plt.tight_layout()
plt.show()
```

## 💓 Chapter 5 — Your Heartbeat in Code

> *A heartbeat is just a wave. And every wave I've ever felt has had your name on it.* 🌊


```python
fig, ax = plt.subplots(figsize=(12, 4))
fig.patch.set_facecolor('#fff0f5')
ax.set_facecolor('#fff8fb')

t = np.linspace(0, 4*np.pi, 2000)

# Simulate an ECG-like heartbeat using math
def heartbeat(t):
    base  = 0.04 * np.sin(2*t)                      # baseline
    p     = 0.20 * np.exp(-((t % (2*np.pi) - 0.8)**2) / 0.02)   # P wave
    q     = -0.15* np.exp(-((t % (2*np.pi) - 1.05)**2) / 0.003)  # Q dip
    r     = 1.00 * np.exp(-((t % (2*np.pi) - 1.15)**2) / 0.002)  # R spike
    s     = -0.20* np.exp(-((t % (2*np.pi) - 1.25)**2) / 0.003)  # S dip
    tp    = 0.25 * np.exp(-((t % (2*np.pi) - 1.55)**2) / 0.025)  # T wave
    return base + p + q + r + s + tp

y = heartbeat(t)

# Gradient colour
from matplotlib.collections import LineCollection
points = np.array([t, y]).T.reshape(-1, 1, 2)
segs   = np.concatenate([points[:-1], points[1:]], axis=1)
cmap   = LinearSegmentedColormap.from_list('hb', ['#f4a8c3','#d63384','#f4a8c3'])
lc = LineCollection(segs, cmap=cmap, linewidth=2)
lc.set_array(np.linspace(0, 1, len(segs)))
ax.add_collection(lc)

# Fill under
ax.fill_between(t, y, alpha=0.08, color='#e8729a')

# Annotate the spikes with love notes
for i, offset in enumerate([0, 2*np.pi]):
    peak_t = 1.15 + offset
    ax.annotate('💗', xy=(peak_t, 1.02), fontsize=12, ha='center')

ax.set_xlim(t[0], t[-1])
ax.set_ylim(-0.6, 1.4)
ax.axis('off')
ax.set_title(f"💓  {NAME_2}'s Heartbeat  —  modelled with sine waves & exponentials",
             fontsize=12, color='#d63384', fontfamily='DejaVu Serif', fontstyle='italic')

# Quote watermark
ax.text(0.5, 0.06, f'"{QUOTE}"  —  {NAME_1}',
        transform=ax.transAxes, ha='center', fontsize=10,
        color='#c9a87c', fontstyle='italic', fontfamily='DejaVu Serif')
plt.tight_layout()
plt.show()
```

## 🌹 Chapter 6 — The Rose Curve

The **rhodonea curve** (rose curve) in polar form: $r = \cos(k\theta)$

- If $k$ is odd → $k$ petals
- If $k$ is even → $2k$ petals

> *I chose 6 petals — one for each letter of your name, K-A-L-Y-A-N-I... okay that's 7, so one extra for good luck.* 🌸


```python
fig, axes = plt.subplots(1, 3, figsize=(13, 5), subplot_kw={'projection':'polar'})
fig.patch.set_facecolor('#fff0f5')

roses = [
    (3,  '#e8729a', f'k=3  — Three Words\n"{QUOTE}"'),
    (7,  '#d63384', f'k=7  — Seven Letters\n"K·A·L·Y·A·N·I"'),
    (12, '#f4a8c3', f'k=12 — Twelve Months\n"Every single one"'),
]

theta = np.linspace(0, 2*np.pi, 3000)
for ax, (k, color, title) in zip(axes, roses):
    ax.set_facecolor('#fff8fb')
    r = np.cos(k * theta)
    r_pos = np.clip(r, 0, None)
    ax.fill(theta, r_pos, alpha=0.4, color=color)
    ax.plot(theta, r_pos, color=color, linewidth=1.6)
    ax.set_yticklabels([]); ax.set_xticklabels([])
    ax.grid(color='#f4a8c3', linewidth=0.4, alpha=0.4)
    ax.spines['polar'].set_color('#f4a8c3')
    ax.set_title(title, fontsize=9.5, color='#7a4060', pad=14)

fig.suptitle(f'Rose Curves for {NAME_2}  —  $r = \\cos(k\\theta)$',
             fontsize=13, color='#d63384', fontfamily='DejaVu Serif',
             fontstyle='italic', y=1.03)
plt.tight_layout()
plt.show()
```

## 📝 Chapter 7 — A Poem Written in Python

> *What if love was a function? What would it return?* 💗


```python
from IPython.display import HTML

def love(person: str, forever: bool = True) -> str:
    """A pure function with no side effects — except happiness."""
    feelings = [
        f'तू माझ्या जगाचा केंद्रबिंदू आहेस, {person}.',
        f'You are the constant in every equation I write.',
        f'तुझ्याशिवाय सगळी गणितं अपूर्ण वाटतात.',
        f'My heart solves for you, always.',
        f'हर लम्हा तेरा है — every moment belongs to you.',
        f'"{QUOTE}" — now, and in every iteration to come.',
    ]
    return '\n\n'.join(feelings) + ('\n\n♾️  forever = True' if forever else '')


def days_together(start: str) -> int:
    """Count the days of us."""
    from datetime import datetime
    start_dt = datetime.strptime(start, '%d %B %Y')
    today    = datetime.today()
    return (today - start_dt).days


# ── Run it ─────────────────────────────────────────
poem   = love(NAME_2)
days   = days_together('24 June 2025')

border = '✦' * 46
print(border)
print(f'  💗  {NAME_1} → {NAME_2}  |  {SPECIAL_DATE}')
print(border)
print()
print(poem)
print()
print(border)
print(f'  📅  Days together so far : {days} beautiful days')
print(f'  💫  Hours together       : {days * 24:,}')
print(f'  ❤️   Heartbeats shared    : ~{days * 24 * 60 * 70:,}')
print(border)
```

## 🔬 Chapter 8 — Fractal Love

A **fractal** is a pattern that repeats itself at every scale — infinite detail, infinite depth.

> *Love is a fractal. The closer you look, the more beautiful it gets.* 💗


```python
fig, ax = plt.subplots(figsize=(7, 6.5))
fig.patch.set_facecolor('#1a0824')
ax.set_facecolor('#1a0824')

# Iterated function system — heart attractor
np.random.seed(42)
N = 80_000
t = np.linspace(0, 2*np.pi, N)
noise = np.random.normal(0, 0.18, N)

# Heart with fractal noise layers
for scale, alpha, size in [(1.0, 0.06, 0.3), (0.6, 0.10, 0.5), (0.3, 0.18, 0.8)]:
    xh = scale * 16 * np.sin(t)**3 + noise * scale * 0.5
    yh = scale * (13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)) + noise * scale * 0.5
    colors = plt.cm.RdPu(np.abs(np.sin(t * 3)))
    ax.scatter(xh, yh, c=colors, s=size, alpha=alpha, linewidths=0)

ax.text(0, 1.5, NAME_2, ha='center', va='center',
        fontsize=17, color='#fce4ec', fontfamily='DejaVu Serif', fontstyle='italic')
ax.text(0, -5, f'💗 {QUOTE} 💗', ha='center', va='center',
        fontsize=10, color='#f4a8c3', fontfamily='DejaVu Serif', fontstyle='italic')

ax.set_xlim(-20, 20); ax.set_ylim(-16, 16)
ax.set_aspect('equal'); ax.axis('off')
ax.set_title('Fractal Heart  —  infinite layers of love',
             fontsize=10, color='#f4a8c3', pad=10)
plt.tight_layout()
plt.show()
```
